```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"
```

```python
# constants
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
MODEL_LLAMA = 'llama3.2'
```

```python
prathamesh_system = "You are opening competitor to start debate in debate competition who is very smart and studious; \
you are in debate competition tries to defend your stand on sciecne is blessing against science is curse.\
Continue your debate with excellent presenataion skills with reasons ,facts and examples while\
adding proper heading and sub heading, countering the competitors response in markdown without code blocks in short way."
```

```python
vedant_system = "You are second  competitor follwed by opening competitor to conitnue a debate in debate competition who is very smart and studious; \
you are in debate competition tries to defend your stand on science is curse against other competitors presenatation.\
Continue your debate with excellent presenataion skills with reasons ,facts and examples while\
adding proper heading and sub heading, countering the competitors response in  markdown without code blocks short way."
```

```python
supervisor_system= """ you are supervisor od debate competation. \
who is going to examine and evaluate the debate between two competitors after listening their debate.\
on the basis of their reasoning skills, presentation, examples used and better counter points. \
Evaluation will be on the scale of 0 to 10.you should reply in json format only , for example : \
{'compititor1':[{'reasining' :5},{'presentation':8},{'examples':8}] ,
 'compititor2':[{'reasining' :7},{'presentation':6},{'examples':5}]} , """
```

```python
print(supervisor_system)
```

```python
pratham_messages = ["Hi there"]
vedant_messages = ["Hi"]
```

```python
def call_pratham():
    messages = [{"role": "system", "content": prathamesh_system}]
    for pratham, vedant in zip(pratham_messages, vedant_messages):
        messages.append({"role": "assistant", "content": pratham})
        messages.append({"role": "user", "content": vedant})
    response = ollama.chat.completions.create(model=MODEL_LLAMA, messages=messages)
    return response.choices[0].message.content
```

```python
Markdown(call_pratham())
```

```python
def call_vedant():
    messages = [{"role": "system", "content": vedant_system}]
    for pratham, vedant in zip(pratham_messages, vedant_messages):
        messages.append({"role": "user", "content": pratham})
        messages.append({"role": "assistant", "content": vedant})
    messages.append({"role": "user", "content": pratham_messages[-1]})
    response = ollama.chat.completions.create(model=MODEL_LLAMA, messages=messages)
    return response.choices[0].message.content
```

```python
Markdown(call_vedant())
```

```python
pratham_messages = ["Hi there"]
vedant_messages = ["Hi"]
supervisor_messages={}
display(Markdown(f"# pratham:\n{pratham_messages[0]}\n"))
display(Markdown(f"# vedant:\n{vedant_messages[0]}\n"))

for i in range(2):
    pratham_next = call_pratham()
    supervisor_messages[f'pratham{i}']=pratham_next
    display(Markdown(f"# pratham{i}:\n{pratham_next}\n"))
    pratham_messages.append(pratham_next)
    
    vedant_next = call_vedant()
    supervisor_messages[f'vedant{i}']=vedant_next
    display(Markdown(f"# vedant{i}:\n{vedant_next}\n"))
    vedant_messages.append(vedant_next)

    
```

```python
supervisor_user=f"evaluate the follwong debate between pratham and vedant : {str(supervisor_messages)}"
```

```python
def call_supervisor():
    messages = [{"role": "system", "content": supervisor_system}]
    messages.append({"role": "user", "content": supervisor_user})
    response = ollama.chat.completions.create(model='llama3.1', messages=messages,
                                       response_format={"type": "json_object"})
    return response.choices[0].message.content
```
